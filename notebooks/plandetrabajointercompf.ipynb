# Plan de Trabajo

## Introducción

El presente documento detalla la estrategia analítica para abordar uno de los desafíos más críticos de Interconnect: la retención proactiva de clientes. A través de un enfoque de Machine Learning, nuestro objetivo no es solo predecir con alta precisión (apuntando a un AUC-ROC $\ge$ 0.88) qué usuarios tienen probabilidad de cancelar sus servicios, sino también comprender los factores que motivan esa decisión. Este plan de acción general traza la ruta metodológica desde la exploración inicial de los datos hasta la entrega de un sistema predictivo robusto. El resultado final permitirá al equipo de marketing actuar de manera anticipada, optimizando la asignación de recursos y códigos promocionales para retener a los clientes de mayor valor.

#### Paso 0: Carga e Inspección Preliminar de los Datos

Antes de cualquier manipulación, se cargarán los 4 archivos CSV (*contract, personal, internet, phone*) para visualizar su estructura real. Esta inspección inicial ya nos permitió detectar requerimientos críticos:

* La variable TotalCharges viene en formato de texto debido a espacios en blanco ocultos que deben ser corregidos.
* No todos los clientes tienen todos los servicios (*ej. algunos solo tienen internet, no teléfono*), por lo que la unificación de tablas requerirá cruces conservadores (*Left Joins*) para no perder clientes.
* La variable objetivo (*fuga*) debe ser extraída lógicamente de la columna *EndDate*.

### Paso 1: Integración y Limpieza Definitiva

Con el diagnóstico del Paso 0, se unificará la información en una única base de datos central. Se forzará la corrección de formatos a numéricos/fechas y se rellenarán los valores nulos generados por la falta de servicios con la etiqueta adecuada (*"No"*).

### Paso 2: Análisis Exploratorio de Datos (EDA)

Se explorarán los datos de manera visual y estadística para comprender el comportamiento de los clientes. Identificaremos qué factores (*como el cargo mensual alto, la baja antigüedad o los contratos mes a mes*) están más fuertemente relacionados con el abandono.

### Paso 3: Ingeniería de Características

A partir de la información limpia, se crearán nuevas variables que ayuden al algoritmo (ej. calcular la antigüedad exacta del cliente desde BeginDate hasta la fecha de corte en febrero 2020, o sumar sus servicios de protección). Finalmente, los datos se traducirán a un formato numérico estándar (One-Hot Encoding).

### Paso 4: Entrenamiento de Modelos Base

La base de datos se dividirá en datos de entrenamiento y prueba (Train/Test). Se entrenará un modelo estadístico inicial (Regresión Logística) aplicando técnicas de peso balanceado (class_weight), considerando que la gran mayoría de los clientes actuales no han cancelado su servicio.

### Paso 5: Modelado Avanzado y Optimización

Se entrenarán algoritmos basados en árboles de decisión de alto rendimiento (LightGBM o XGBoost). Se realizarán pruebas automatizadas (Tuning de hiperparámetros) ajustando sus configuraciones internas hasta encontrar la combinación que maximice nuestra métrica objetivo (AUC-ROC $\ge$ 0.88).

### Paso 6: Evaluación y Recomendaciones de Negocio

El modelo final será puesto a prueba con datos nunca antes vistos. Se analizará qué características tuvieron mayor peso en las decisiones del algoritmo y se redactará un informe final con recomendaciones estratégicas para las campañas operativas de retención de Interconnect.

## Conclusión

En resumen, esta hoja de ruta asegura que cada esfuerzo técnico esté estrictamente alineado con las necesidades comerciales de Interconnect. Al completar estas fases, no solo entregaremos un algoritmo predictivo, sino una herramienta estratégica que transformará datos históricos en campañas de retención eficaces y medibles.
